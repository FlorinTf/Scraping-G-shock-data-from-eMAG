```python
import pandas as pd
```

```python
gs=pd.read_csv('g_shock.csv')
gs
```

```python
#The script collected duplicate items so I drop the duplicated.
gs = gs.drop_duplicates()
gs
```

```python
gs.reset_index(inplace=True)
gs
```

```python
#The price ar in RON so I make new column with price in Euro
euro_price = []
for i in range(len(gs)):
    price = gs['Pret'][i]//4.94
    euro_price.append(int(price))
gs.iloc[0]['Price EU']=euro_price
gs
```
